```python
import numpy as np 
import h5py
import matplotlib.pyplot as plt
from scipy.signal import spectrogram as spec
```

```python
import os
from pathlib import Path
```

```python
folder_name = 'waveforms_160_40Twz_5lambda_susc-meas'

file_name = Path(folder_name,'static.h5')
hf = h5py.File(file_name, 'r')
a = hf.get('A')

file_name = Path(folder_name,'sweep_Lo16.h5')
hf = h5py.File(file_name, 'r')
ab = hf.get('A')

file_name = Path(folder_name,'static_shifted_Lo16.h5')
hf = h5py.File(file_name, 'r')
b = hf.get('A')

file_name = Path(folder_name,'sweep_back_1Lo16.h5')
hf = h5py.File(file_name, 'r')
ba = hf.get('A')


wf = np.concatenate((a[:],a[:],a[:],a[:], ab[:], b[:], ba[:], a[:],a[:],a[:],a[:]))

data = wf
timeperseg = 100 #us
samplerate = 1e3 #1e3 S / us = 1Gs/s 
powerSpectrum, freqenciesFound, time, imageAxis = plt.specgram(data,\
                                                               Fs=samplerate,\
                                                               NFFT=int(timeperseg*samplerate),\
                                                               noverlap=int(.95*timeperseg*samplerate),\
                                                               scale='dB',cmap='PRGn')

plt.xlabel('Time')

plt.ylabel('Frequency')

plt.ylim([85, 125])
plt.xlim([50,200])
plt.colorbar()
plt.clim(-150,130)
# plt.clim(0,1e8)
```
