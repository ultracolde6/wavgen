```python
wf_test = np.concatenate((a[-160000:], ab[:], a[-160000:]))
```

```python
data = wf_test
timeperseg = 100 #us
samplerate = 1e3 #1e3 S / us = 1Gs/s 
powerSpectrum, freqenciesFound, time, imageAxis = plt.specgram(data,\
                                                               Fs=samplerate,\
                                                               NFFT=int(timeperseg*samplerate),\
                                                               noverlap=int(.95*timeperseg*samplerate),\
                                                               scale='dB',cmap='PRGn')

plt.xlabel('Time')

plt.ylabel('Frequency')

plt.ylim([92,102])
plt.colorbar()
plt.clim(-100,130)
# plt.clim(0,2e8)
```
